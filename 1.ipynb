```python
import librosa
import numpy as np
import matplotlib.pyplot as plt

for i in range(1,11):
    
    audio_file = f"{i}.mp3"
    audio_signal, sampling_rate = librosa.load(audio_file)

    duration = librosa.get_duration(y=audio_signal, sr=sampling_rate)

    time = np.arange(0, len(audio_signal)) / sampling_rate

    plt.figure(figsize=(10, 4))
    plt.plot(time, audio_signal, linewidth=0.5, color='b')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.title(f'Waveform of Audio Signal of Sample-{i}')
    plt.grid(True)

    plt.xlim(0, duration)

    plt.savefig(f'audio_waveform_{i}.png')

    reversed_signal = 1 - audio_signal

    plt.figure(figsize=(10, 4))
    plt.plot(time, reversed_signal, linewidth=0.5, color='r')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.title(f'Reversed Waveform of Audio Signal Sample-{i}')
    plt.grid(True)

    plt.xlim(0, duration)

    plt.savefig(f'reversed_audio_waveform_{i}.png')

    plt.show()
    import numpy as np
    import matplotlib.pyplot as plt

    start_time = 2
    end_time = 2.1

    start_sample = int(start_time * sampling_rate)
    end_sample = int(end_time * sampling_rate)

    portion_signal = audio_signal[start_sample:end_sample]
    portion_time = np.linspace(start_time, end_time, len(portion_signal))

    plt.figure(figsize=(10, 4))

    # Plot original waveform
    plt.plot(portion_time, portion_signal, linewidth=0.5, color='b', label='Original')


    # Plot reversed waveform
    reversed_portion_signal = 1 - portion_signal
    plt.plot(portion_time, reversed_portion_signal, linewidth=0.5, color='c', label='Reversed')

    # Merge signals
    merged_signal = portion_signal + reversed_portion_signal
    plt.plot(portion_time, merged_signal, linewidth=1, color='g', label='Merged')

    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.title(f'Waveforms of Portion (2-2.1 sec) of Audio Signal Sample-{i}')
    plt.grid(True)
    plt.legend()
    plt.savefig(f'merged_portion_audio_waveforms_{i}.png')
    plt.show()
```
